```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

# Load the dataset (assuming it's already loaded as 'df')
# If not, uncomment the following lines and replace with your actual path

def leading_digit_distribution(series):
    """Calculates the leading digit distribution of a numerical series."""
    leading_digits = []
    for value in series:
        if not pd.isna(value) and value != 0 :
          leading_digits.append(int(str(int(abs(value)))[0]))

    counts = pd.Series(leading_digits).value_counts(normalize=True, sort=False)
    return counts


def benfords_law(digits):
    """Returns Benford's Law distribution for digits 1-9."""
    benford_dist = {}
    for digit in digits:
        benford_dist[digit] = np.log10(1 + (1 / digit))
    return benford_dist


numerical_cols = df.select_dtypes(include=np.number).columns


for col in numerical_cols:
    observed_dist = leading_digit_distribution(df[col])


    expected_dist = benfords_law(range(1,10))
    expected_dist = pd.Series(expected_dist)

    # Visualize the results
    fig = go.Figure()
    fig.add_trace(go.Bar(x=observed_dist.index, y=observed_dist.values, name='Observed'))
    fig.add_trace(go.Bar(x=expected_dist.index, y=expected_dist.values, name='Benford\'s Law'))
    fig.update_layout(title=f'Leading Digit Distribution for {col}', xaxis_title='Leading Digit', yaxis_title='Frequency')
    fig.show()

    # Compare observed and expected distributions (e.g., using Chi-squared test)
    # ... (add Chi-squared or other statistical test here to quantify the difference)
    # Print or store the results of the statistical tests

    # Interpretation:
    print(f"\nAnalysis for column '{col}':")
    # If you performed a Chi-squared test, print the p-value and significance level
    # Example (replace with actual p-value):
    # p_value = 0.05
    # if p_value < 0.05:
        # print("The observed distribution significantly deviates from Benford's Law.")
    # else:
        # print("The observed distribution does not significantly deviate from Benford's Law.")
    # Discuss any observed patterns, deviations, or anomalies in the data.

```

```python
# Extract the first significant digit from each numeric entry.
# Create a new column for the leading digit. and visualise
# Ensure data integrity (e.g., filter out non-positive or malformed values).

# Assuming df is already loaded as in your provided code

def extract_leading_digit(value):
    """Extracts the first significant digit from a numeric value."""
    if pd.isna(value) or value == 0:
        return None  # Handle missing or zero values
    try:
      abs_value = abs(value)
      return int(str(int(abs_value))[0])
    except (ValueError, IndexError, TypeError):
      return None



for col in numerical_cols:
    df[f'{col}_leading_digit'] = df[col].apply(extract_leading_digit)

# Visualize the leading digit distribution for each column
for col in numerical_cols:
    leading_digit_counts = df[f'{col}_leading_digit'].value_counts(normalize=True, sort=False).sort_index()

    fig = go.Figure()
    fig.add_trace(go.Bar(x=leading_digit_counts.index, y=leading_digit_counts.values, name='Leading Digit Frequency'))

    fig.update_layout(
        title=f'Leading Digit Distribution for {col}',
        xaxis_title='Leading Digit',
        yaxis_title='Frequency',
    )
    fig.show()
```
